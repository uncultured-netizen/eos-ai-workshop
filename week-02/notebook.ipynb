```python
# Arithmetic Operators in Python

a = 10
b = 3

# Addition
print("Addition:", a + b)       # 10 + 3 = 13

# Subtraction
print("Subtraction:", a - b)    # 10 - 3 = 7

# Multiplication
print("Multiplication:", a * b) # 10 * 3 = 30

# Division (float result)
print("Division:", a / b)       # 10 / 3 = 3.333...

# Floor Division (integer result)
print("Floor Division:", a // b) # 10 // 3 = 3

# Modulus (remainder)
print("Modulus:", a % b)        # 10 % 3 = 1

# Exponentiation (power)
print("Exponent:", a ** b)      # 10^3 = 1000
```

```python
def calculate_percentage(marks_list):
    # 1. Add all marks
    total = sum(marks_list)

    # 2. Divide by max possible (each out of 100)
    percentage = total / (len(marks_list) * 100) * 100

    # 3. Round to 2 decimal places
    return round(percentage, 2)


# Use it
marks = [74, 65, 76, 67, 56]
result = calculate_percentage(marks)
print(f"I scored: {result}%")
if result>=85:
  print(f"A grade!")
else:
  print(f"Not an A grade!")
```

```python
import numpy as np

scores = np.array([88, 92, 76, 85, 90])

print(scores.mean())    # 86.2      ← average
print(scores.std())     # 5.56      ← standard deviation (spread)
print(scores.max())     # 92
print(scores.min())     # 76
print(scores.sum())     # 431

# Where's the max/min?
print(np.argmax(scores))  # 1       ← index of the top scorer (92)
print(np.argmin(scores))  # 2       ← index of the lowest (76)

# Sort
print(np.sort(scores))    # [76 85 88 90 92]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create a 16×16 canvas of all ones (white)
canvas = np.ones((16, 16), dtype=int)

# 2. Set some cells to 0 (black) to "draw"
canvas[5, 3] = 0        # single pixel at row 5, col 3
canvas[4,10] = 0        # single pixel at row 4, col 10
canvas[9, 4:11] = 0     # horizontal line at row 9, from col 4 to 11
canvas[3, 3:6] = 0
canvas[3, 3:6] = 0
canvas[10, 5:10] = 0
canvas[2, 9:13] = 0
canvas[1, 1] = 0
canvas[1, 2] = 0
canvas[2, 1] = 0
canvas[1, 14] = 0
canvas[1, 13] = 0
canvas[2, 14] = 0
canvas[14, 1] = 0
canvas[14, 2] = 0
canvas[13, 1] = 0
canvas[14, 14] = 0
canvas[13, 14] = 0
canvas[14, 13] = 0
# 3. Use slicing to add a border
canvas[0, :] = 0
canvas[-1, :] = 0
canvas[:, 0] = 0
canvas[:, -1] = 0

# 4. Display
plt.imshow(canvas, cmap='gray')
plt.axis('off')
plt.show()
```
